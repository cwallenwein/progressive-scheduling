```python
import torch
from torch.optim import SGD
from progressive_scheduling.visualize import plot_lr_scheduler
from progressive_scheduling import OneCycleLR


def create_optimizer():
    model = torch.nn.Linear(1, 1)
    optimizer = SGD(model.parameters(), lr=0.1)
    return optimizer


optimizer = create_optimizer()
pytorch_scheduler = torch.optim.lr_scheduler.OneCycleLR(optimizer, max_lr=0.1, steps_per_epoch=100, epochs=10)

optimizer = create_optimizer()
progressive_scheduler = OneCycleLR(optimizer, max_lr=0.12)

plot_lr_scheduler(pytorch_scheduler=pytorch_scheduler, progressive_scheduler=progressive_scheduler, num_steps=1000)
```
